```python
import pandas as pd
MULHERES_BR_1 = pd.read_excel('../../Datasets/MULHERES_BR-1.xlsx', index_col=1)
# resetando index
MULHERES_BR_1.reset_index(inplace=True)
MULHERES_BR_1
```

```python
import pandas as pd
TOTAL_BR = pd.read_excel('../../Datasets/TOTAL_BR.xlsx', index_col=1)
# nomeando coluna Unnamed: 0 para Região
TOTAL_BR.rename(columns={'Unnamed: 0':'Região'}, inplace=True)
# resetando index
TOTAL_BR.reset_index(inplace=True)  
#trocar ordem das colunas total e região


# criar dicionário chave-valor de cada total por região
total_regiao = {}
for i in range(len(TOTAL_BR)):
    total_regiao[TOTAL_BR['Total'][i]] = TOTAL_BR['Região'][i]  
TOTAL_BR
```

```python
# criar novo DF com somatório das colunas de faixa etária para homens
total_homens = HOMENS_BR_1.sum(axis=1)  
total_mulheres = MULHERES_BR_1.sum(axis=1)
df_total = pd.DataFrame(total_homens, columns=['Homens'])
df_total['Mulheres'] = total_mulheres
df_total['Total'] = df_total['Homens'] + df_total['Mulheres']

# criar coluna região obtendo valores do dicionário total_regiao

df_total['Região'] = df_total['Total'].map(total_regiao)

df_total
```

```python
import os
import boto3
import ssl
import pandas as pd

class AmazonS3:
    def __init__(self):
        ssl._create_default_https_context = ssl._create_unverified_context
        self.bucket = None
        self.bucket_name = 'data-lake-turma6-puc'
        self.aws_access_key_id = ''
        self.aws_secret_access_key = ''
        self.region_name = 'sa-east-1'
        self.resource = 's3'

    def _connect_s3(self):
        try:
            session = boto3.Session(
                aws_access_key_id=self.aws_access_key_id,
                aws_secret_access_key=self.aws_secret_access_key,
                region_name=self.region_name
            )
            s3 = session.resource(self.resource)
            self.bucket = s3.Bucket(self.bucket_name)
        except Exception as e:
            print(e)

    def _put_object_bucket(self, local_path, amazon_path=None):
        if not amazon_path:
            amazon_path = "resultado_tarefa_IV/"
        amazon_destiny = os.path.join(amazon_path, os.path.basename(local_path))
        try:
            with open(local_path, 'rb') as data:
                self.bucket.put_object(Key=amazon_destiny, Body=data)
        except Exception as e:
            print(e)
            return None
        url = f'https://{self.bucket_name}.s3.amazonaws.com/{amazon_destiny}'
        print(url)
        return url

    def upload_dataframe(self, df, file_name):
        local_path = f"{file_name}.csv"
        df.to_csv(local_path, index=False)
        self._connect_s3()
        return self._put_object_bucket(local_path)


# Salvar o dataframe como CSV localmente
file_name = "resultado_tarefa_IV_Luciano"
df_total.to_csv(f"{file_name}.csv", index=False)

# Instancia e uso da classe AmazonS3
AS3 = AmazonS3()
upload_url = AS3.upload_dataframe(df_total, file_name)
print(f"Arquivo enviado para: {upload_url}")
```
