### Preparação dos dados

```python
import pandas as pd
df = pd.read_excel('../../Datasets/tabela9514_UF_MUN_MG.xlsx')
df
```

```python
# Identificar os índices das colunas para 'Homens' e 'Mulheres'
# Identificar os índices das colunas para 'Homens' e 'Mulheres'
index_homens_pos = df.columns.get_loc((df.loc[3] == 'Homens').idxmax())
index_mulheres_pos = df.columns.get_loc((df.loc[3] == 'Mulheres').idxmax())
print(index_mulheres_pos)

# Selecionar as colunas de 'Homens' até uma posição antes de 'Mulheres'
df_homens = df.iloc[:-1, index_homens_pos:index_mulheres_pos]
df_mulheres = df.iloc[:-1, index_mulheres_pos:]

df_homens = df_homens.iloc[4:]
df_mulheres = df_mulheres.iloc[4:]
# Tornar a primeira linha os nomes das colunas
df_homens.columns = df_homens.iloc[0]
df_mulheres.columns = df_mulheres.iloc[0]
df_homens = df_homens[1:]
df_mulheres = df_mulheres[1:]

# Redefinir os índices do DataFrame
df_homens.reset_index(drop=True, inplace=True)
df_mulheres.reset_index(drop=True, inplace=True)       
```

```python
# preparar a coluna cidade para ser o index
df_cidades = df.iloc[5:-1,[0]]
df_cidades.columns = ['Cidade'] 
```

```python
df_cidades
```

```python
# Adicionar a coluna 'Cidade' como índice
df_homens.set_index(df_cidades['Cidade'], inplace=True)
df_mulheres.set_index(df_cidades['Cidade'], inplace=True)
```

```python
df_homens
```

```python
df_mulheres
```

```python
import numpy as np

# Substituir valores nulos por '-'
df_homens = df_homens.replace(np.nan, '-')
df_mulheres = df_mulheres.replace(np.nan, '-')  

# Substituir valores '-' por 0
df_homens = df_homens.replace('-', 0)
df_mulheres = df_mulheres.replace('-', 0)

# Converter os valores para float
df_homens = df_homens.astype(float) 
df_mulheres = df_mulheres.astype(float) 
```

Ajustando colunas úteis

```python
# removendo coluna total
df_homens = df_homens.drop(columns=['Total'])
df_mulheres = df_mulheres.drop(columns=['Total'])

# remover primeira linha
df_homens = df_homens.iloc[1:]  
df_mulheres = df_mulheres.iloc[1:]  
```

### a) Faça um programa em python que identifique qual a faixa etaria possui mais pessoas entre homens e mulheres.

```python
# Calcular a soma de cada faixa etária para homens e mulheres
sum_homens = df_homens.sum()
sum_mulheres = df_mulheres.sum()

# Calcular a faixa etária com mais pessoas para homens e mulheres   
max_homens = sum_homens.idxmax()    
max_mulheres = sum_mulheres.idxmax()

print(f'A faixa etária com mais pessoas entre os homens é: {max_homens}')
print(f'A faixa etária com mais pessoas entre as mulheres é: {max_mulheres}')   
```

### b) Faça um programa python que com base na sua faixa etária e sexo identifique o municipio com mais pessoas.        

```python
# eu tenho 26 anos então a faixa etária é 25 a 29 anos
# Calcular a cidade com mais pessoas para homens na faixa etária de 25 a 29 anos

print(df_homens['25 a 29 anos'].idxmax())
```

### c) Faça um programa python que com base na sua faixa etária e sexo identifique o municipio com menos pessoas.   

```python
print(df_homens['25 a 29 anos'].idxmin())
```

### d) Faça um programa em python que com base no seu municipio de nascimento identifique a maior faixa etária em anos entre homens e mulheres.

```python
# meu municipio é Buritis (MG)
# Calcular a faixa etária com mais pessoas para homens e mulheres em Buritis
print(df_homens.loc['Buritis (MG)'].idxmax())
print(df_mulheres.loc['Buritis (MG)'].idxmax())  
```

### e) Faça um programa em python que gere dois arquivos um em .xlsx e o outro .json com o resultado das questões acima e salve no  data-lake-turma6-puc/Prova/ com seu nome a extensão de cada arquivo.    

```python
# salvar os resultados em um arquivo .xlsx

df_result = pd.DataFrame({
    'Faixa etária com mais pessoas entre os homens': [max_homens],
    'Faixa etária com mais pessoas entre as mulheres': [max_mulheres],
    'Cidade com mais pessoas para homens na faixa etária de 25 a 29 anos': [df_homens['25 a 29 anos'].idxmax()],
    'Cidade com menos pessoas para homens na faixa etária de 25 a 29 anos': [df_homens['25 a 29 anos'].idxmin()],
    'Faixa etária com mais pessoas para homens em Buritis (MG)': [df_homens.loc['Buritis (MG)'].idxmax()],
    'Faixa etária com mais pessoas para mulheres em Buritis (MG)': [df_mulheres.loc['Buritis (MG)'].idxmax()]
})  
df_result
```

```python
# salvar os resultados em um arquivo .xlsx
df_result.to_excel('Luciano_Contri.xlsx', index=False)

# salvar os resultados em um arquivo .json
df_result.to_json('Luciano_Contri.json', orient='records')
```
